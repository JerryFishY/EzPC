## Import

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

scale = 23
```

## Define model

```python
class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        self.conv = nn.Conv2d(4, 4, 3, groups=4, bias=False, padding=1)
        
    def forward(self, x):
        return self.conv(x)
```

## Create dummy input

```python
torch.manual_seed(0)

model = Net()
model.eval()
dummy_input = torch.randn(1, 4, 10, 10)
```

## Export

```python
torch.onnx.export(
    model,
    dummy_input,
    "conv3_group.onnx",
    verbose=False,
    input_names=['input_names'],
    output_names=['output_names'],
    export_params=True
)
```

## Generate input

```python
with open("Inputs/conv3_group_input.inp", 'w') as f :
    for p in dummy_input.flatten() :
        fixed = int(p*(2**scale))
        # print(fixed)
        f.write(f"{fixed}\n")
        
    f.close()
```

```python
out = model(dummy_input)
for p in out.flatten() :
    fixed = int(p*(2**scale))
    print(fixed)
```
